```python
stats['min_age'] = stats['min_age'].apply(lambda x: 19 if x <= 19 else x)
plt.figure(figsize=(10,6))
sns.set_context('notebook')
plt.xticks(ticks = [x for x in range(18,40)])
plt.title("PER by Age, Segmented by Age in First Season")
plt.xlim((19,40))

for i in range(19, 24):
    temp_df = stats[stats['min_age'] == i][['Age', 'PER']]
    temp_df = temp_df.groupby(by = 'Age').agg(['mean', 'count'])
    temp_df.columns = [' '.join(col).strip() for col in temp_df.columns.values]
    temp_df = temp_df[temp_df['PER count'] > 10]['PER mean']
    sns.lineplot(x = temp_df.index, y = temp_df, label = i)
    sns.scatterplot(x = temp_df.index, y = temp_df);
```

```python
temp_df = stats[np.array(stats['min_age'] <= 23) & np.array(stats['Age'] <= 28) &
                  np.array(stats['Age'] >= 24)]

plt.figure(figsize=(10,6))
plt.ylim((0,35))
sns.set_context('notebook') # set to 'talk' for bigger font and lines
sns.boxplot(x = temp_df['min_age'], y = temp_df['PER'])
plt.title('PER in Prime Years (24-28) by Entry Age');
```
